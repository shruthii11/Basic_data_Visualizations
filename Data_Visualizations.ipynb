### In Matplotlib, a visualization refers to the graphical representation of data, patterns, or relationships using various plot types and elements. Matplotlib is a powerful library for creating a wide range of visualizations in Python.

Here are some common types of visualizations that can be created using Matplotlib:

1. **Line Plot**: Displays data points connected by lines, commonly used for time series or continuous data.

2. **Scatter Plot**: Represents individual data points as dots, useful for showing relationships between two variables.

3. **Bar Plot**: Displays rectangular bars with lengths proportional to the values they represent, useful for comparing discrete data.

4. **Histogram**: Shows the distribution of data by dividing it into intervals (bins) and displaying the frequency of observations in each bin.

5. **Pie Chart**: Represents data as slices of a circular pie, showing the composition of a whole in terms of its parts.

6. **Box Plot (Box-and-Whisker Plot)**: Provides a summary of the distribution of a dataset by showing the median, quartiles, and potential outliers.

7. **Heatmap**: Displays a matrix where individual values are represented as colors, useful for showing correlations or relationships between two variables.

8. **Stacked Bar Plot**: Combines multiple bar plots to display the total and individual contributions of each category.

10. **Scatter Matrix (Pair Plot)**: Displays scatter plots for pairs of variables in a dataset to show relationships between multiple variables.

These are just some examples of the many types of visualizations you can create using Matplotlib. Matplotlib provides a versatile set of tools to help you effectively visualize your data and communicate insights.

# Pie Chart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['blue', 'green', 'red', 'purple']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

plt.axis('equal')
plt.title('Basic Pie Chart')

plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['blue', 'green', 'red', 'purple']
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')

plt.title('Basic Pie Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['blue', 'green', 'red', 'purple']

# Create a donut chart by creating a white circle at the center
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90,
        wedgeprops={'edgecolor': 'gray', 'linewidth': 2}, pctdistance=0.85)

# Draw a white circle at the center to create the "donut hole"
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

plt.title('Donut Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 28, 30, 15]

plt.figure(figsize=(8, 6))

plt.bar(categories, values, color='green', edgecolor='black', linewidth=1.2)

plt.xlabel('Categories', fontsize=12, fontweight='bold')
plt.ylabel('Values', fontsize=12, fontweight='bold')
plt.title('Customized Bar Chart', fontsize=14, fontweight='bold')

plt.xticks(rotation=40, ha='right')


plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 15]

plt.figure(figsize=(10, 6))

bars = plt.bar(categories, values, color='steelblue', edgecolor='black', linewidth=1.2, alpha=0.7)

plt.xlabel('Categories', fontsize=12, fontweight='bold')
plt.ylabel('Values', fontsize=12, fontweight='bold')
plt.title('Advanced Bar Chart', fontsize=14, fontweight='bold')

plt.xticks(rotation=45, ha='right', fontsize=10)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 2, round(yval, 2), ha='center', fontsize=10)

plt.ylim(0, max(values) + 10)
plt.grid(axis='y', linestyle='-.', alpha=0.7)
plt.show()
```

## Stacked bar plot

```python
import matplotlib.pyplot as plt

a1 = [1, 3, 6, 9]
a2 = [2, 36, 8, 9]
a3 = [23, 45, 78, 59]
a4 = [21, 32, 65, 54]
label = ['arr1', 'arr2', 'arr3', 'arr4']

plt.style.use('seaborn')

plt.stackplot(range(len(a1)), a1, a2, a3, a4, labels=label, alpha=0.7)

plt.legend(loc=(0.4, 0.9))
plt.tight_layout()

plt.xlabel('X-axis')
plt.ylabel('Values')
plt.title('Stacked Plot')

plt.show()
```

# Basic Matplotlib plots

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = 2 * x + 1 + np.random.normal(0, 1, 50)
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 15]
data = np.random.normal(0, 1, 100)
iris = sns.load_dataset("iris")

# Create a figure with subplots
fig, axes = plt.subplots(3, 3, figsize=(10, 10))

# Line Plot
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Line Plot')

# Scatter Plot
axes[0, 1].scatter(x, y)
axes[0, 1].set_title('Scatter Plot')

# Bar Plot
axes[0, 2].bar(categories, values)
axes[0, 2].set_title('Bar Plot')

# Histogram
axes[1, 0].hist(data, bins=10, color='skyblue', edgecolor='black')
axes[1, 0].set_title('Histogram')

# Pie Chart
axes[1, 1].pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
axes[1, 1].set_title('Pie Chart')

# Box Plot
sns.boxplot(data=iris, ax=axes[1, 2])
axes[1, 2].set_title('Box Plot')

# Heatmap
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm', ax=axes[2, 0])
axes[2, 0].set_title('Heatmap')

# Stacked Bar Plot
data = np.array([[10, 20, 15, 5], [15, 10, 5, 20], [5, 25, 10, 10]])
categories = ['Category A', 'Category B', 'Category C', 'Category D']
axes[2, 1].bar(categories, data[0], color='b', label='Group 1')
axes[2, 1].bar(categories, data[1], bottom=data[0], color='g', label='Group 2')
axes[2, 1].bar(categories, data[2], bottom=data[0] + data[1], color='r', label='Group 3')
axes[2, 1].legend()
axes[2, 1].set_title('Stacked Bar Plot')

# Pair Plot (Scatter Matrix)
pair_grid = sns.PairGrid(iris, diag_sharey=False, height=1.5)
pair_grid.map_upper(plt.scatter)
pair_grid.map_lower(sns.kdeplot)
pair_grid.map_diag(sns.kdeplot, lw=3)
pair_grid.add_legend()

plt.tight_layout()
plt.show()

```
