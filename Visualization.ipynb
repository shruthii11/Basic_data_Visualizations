```python
years = [y for y in range(2000,2011)]
apples = [0.895,0.91,0.919,0.926,0.929,0.931,.934,.936,.937,.9375,.939]
oranges = [0.962,.941,.930,.923,.918,.908,.904,.901,.898,.9,.896]
```

```python
plt.plot(years,apples);
plt.plot(years,oranges);
plt.xlabel('years')
plt.ylabel('yield (tons per hectare)')
```

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')
plt.title('crops yields in kanto')
plt.legend(['apples','oranges']);
```

# Line markers

```python
plt.plot(years,apples,marker='1')
plt.plot(years,oranges,marker='2')

plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')

plt.title('crops yields in kanto')
plt.legend(['apples','oranges']);
```

```python
plt.plot(years,apples,marker='s',c='b',ls='-',lw=2,ms=10,mew=3,mec='plum')
plt.plot(years,oranges,marker='s',c='r',ls='--',lw=3,ms=10,alpha=.59,mec='lavender')

plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')

plt.title('Crop yield in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(years,apples,'s-b')
plt.plot(years,oranges,'o--r')
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')

plt.title('Crop yield in Kanto')
plt.legend(['Apples','Oranges']);
```

# Changing the Figure size

# Scatter plot

```python
# load data into a pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique() # to find unique values in a column
```

```python
plt.plot(flowers_df.sepal_length,flowers_df.sepal_width);
```

```python
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width);
```

```python
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width, hue=flowers_df.species,s=60);
```

# Customizing Seaborn figures

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width, hue=flowers_df.species,s=60);
```

# Histogram

```python
plt.title('Distribution sepal Width')
plt.hist(flowers_df.sepal_width);
```

# Controlling size and number of bins

```python
plt.hist(flowers_df.sepal_width,bins=14)
```

```python
import numpy as np
plt.hist(flowers_df.sepal_width,bins=np.arange(2,5,0.25))
```

```python
plt.hist(flowers_df.sepal_width,bins=[1,3,4,4.5]);
```

# Multiple Histograms

```python
setosa_df=flowers_df[flowers_df.species=='setosa']
versicolor_df=flowers_df[flowers_df.species=='versicolor']
virginica_df=flowers_df[flowers_df.species=='verginica']
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
```

```python
plt.title('Distribution of sepal width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True);
plt.legend(['Setosa','Versicolor','Virginica']);
```
